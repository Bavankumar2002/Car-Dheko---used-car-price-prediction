## Car Dheko - used car price Prediction

### Unstructured data convert into structure data


(i) Chennai cars data

```python
import pandas as pd
import ast

# Load the Excel file
file_path = 'chennai_cars (2).xlsx'  # Replace with your file path
xls = pd.ExcelFile(file_path)

# Load the sheet containing unstructured data
df = pd.read_excel(file_path, sheet_name='chennai_cars.csv')

# Function to extract data from JSON-like fields
def extract_info(row):
    # Parse the JSON-like data from string
    car_detail = ast.literal_eval(row['new_car_detail'])
    car_overview = ast.literal_eval(row['new_car_overview'])
    car_feature = ast.literal_eval(row['new_car_feature'])
    car_specs = ast.literal_eval(row['new_car_specs'])

    # Extract relevant fields
    fuel_type = car_detail.get('ft')
    body_type = car_detail.get('bt')
    kilometers_driven = car_detail.get('km')
    car_model = car_detail.get('model')
    car_price = car_detail.get('price')

    registration_year = next((item['value'] for item in car_overview['top'] if item['key'] == 'Registration Year'), None)

    ownership = next((item['value'] for item in car_overview['top'] if item['key'] == 'Ownership'), None)

    features = ', '.join([f['value'] for f in car_feature['top']])

    engine_power = next((item['value'] for item in car_specs['top'] if item['key'] == 'Max Power'), None)
    mileage = next((item['value'] for item in car_specs['top'] if item['key'] == 'Mileage'), None)

    car_link = row['car_links']

    return pd.Series([fuel_type, body_type, kilometers_driven, registration_year, car_model, ownership, features, engine_power, mileage, car_price ,car_link])

# Apply the extraction function to each row in the DataFrame
df_structured_chennai = df.apply(extract_info, axis=1)

# Assign column names to the structured DataFrame
df_structured_chennai.columns = ['Fuel Type', 'Body Type', 'Kilometers Driven', 'Registration Year', 'Car Model', 'Ownership', 'Features', 'Engine Power', 'Mileage', 'Car Price','Car Link']

# Save the structured data to a new Excel file
output_path = 'structured_Chennai_cars_data.csv'
df_structured_chennai.to_csv(output_path, index=False)

print(f"Structured data saved to {output_path}")
```

### Merge all cars data


```python
import pandas as pd

chennai_df = pd.read_csv('structured_Chennai_cars_data.csv')
bangalore_df = pd.read_csv('structured_Bangalore_cars_data.csv')
hyderabad_df = pd.read_csv('structured_hyderabad_cars_data.csv')
delhi_df = pd.read_csv('structured_delhi_cars_data.csv')
kolkata_df = pd.read_csv('structured_kolkata_cars_data.csv')

# Add a new column 'City' and assign the value to all rows

chennai_df['City'] = 'Chennai'
bangalore_df['City'] = 'Bangalore'
delhi_df['City'] = 'Delhi'
hyderabad_df['City'] = 'Hyderabad'
kolkata_df['City'] = 'Kolkata'


# Concatenate all datasets into a single dataset
combined_df = pd.concat([chennai_df,bangalore_df,delhi_df,hyderabad_df,kolkata_df])

# Save the combined dataset to a new CSV file
combined_df.to_csv('cars_details.csv', index=False)

print("Combined dataset saved to 'cars_details.csv'")
```
